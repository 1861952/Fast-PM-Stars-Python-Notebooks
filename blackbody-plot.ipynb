```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.modeling.models import BlackBody
from astropy import units as u
from astropy.visualization import quantity_support

fig, ax = plt.subplots(figsize=(10,6))

temps = np.array([3000, 4000, 5000])
wav = np.arange(1000, 1000000) * u.AA

plt.figure()
with quantity_support():
    for t in temps:
        # print(t)
        bb = BlackBody(temperature=t*u.K)
        flux = bb(wav)
        ax.plot(wav, flux, label=f"$T=$" + str(t) + "K")
        # plt.axvline(bb.nu_max.to(u.AA, equivalencies=u.spectral()).value, ls='--')
ax.set_xscale("log")
ax.set_title("Black Body")
ax.legend()
```

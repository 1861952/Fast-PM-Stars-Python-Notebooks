```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import powerlaw

fig, ax = plt.subplots(1,1)
a = 0.35
mean, var, skew, kurt = powerlaw.stats(a, moments='mvsk')

#  Display pdf
x = np.linspace(powerlaw.ppf(0.1, a), powerlaw.ppf(0.99, a), 100)

ax.plot(x, powerlaw.pdf(x,a), 'r-', lw=5, alpha=0.6, label='powerlaw pdf')

# Freeze distribution
rv = powerlaw(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

# Plot Histogram
r = powerlaw.rvs(a, size=100000)
hist, bins, _ = ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
logbins = np.logspace(np.log10(bins[0]), np.log10(bins[-1]), len(bins))
print(logbins)
# ax.hist(r, density=True, bins=logbins, histtype='step', alpha=0.2 )
ax.set_xlim([x[0], x[-1]])
plt.xscale("log")
plt.yscale("log")
ax.legend(loc='best', frameon=False)
ax.set_xlabel(r"Solar Masses")
ax.set_ylabel(r"$log(N)$")
plt.show()
fig.savefig("pow-law-model.png")
```

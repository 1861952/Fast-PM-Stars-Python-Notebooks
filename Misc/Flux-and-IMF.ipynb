# Flux #

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u


F2_F1_ratio = np.linspace(0.1, 10)
delta_m = -2.5*np.log10(F2_F1_ratio)
    
fig, ax = plt.subplots(figsize=(10, 6))
# mag2 = get_mag(mag1, flux1, flux2)
# plt.xscale("log")
ax.plot(F2_F1_ratio, delta_m)
ax.set_xlabel(r"$\frac{F_2}{F_1}$")
ax.set_ylabel(r"$m_2 - m_1$")
ax.set_title("Apparent Magnitude Difference as a Function of Flux Ratio")
plt.grid()
plt.show()
fig.savefig("mag_flux.png")

```

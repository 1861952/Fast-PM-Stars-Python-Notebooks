```python
import numpy as np
from scipy.stats import norm, powerlognorm
import matplotlib.pyplot as plt
```

# Gaussian Distribution #

```python
fig, ax = plt.subplots()
# calculate first 4 moments
mean, var, skew, kurt = norm.stats(moments='mvsk')

# Display pdf (not for continuous distribution)
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')

# Freeze? distribution 
rv = norm()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

# Check accuracy? of cdf and ppf
vals = norm.ppf([0.001, 0.5, 0.999]) 
# Returns True if two arrays are element-wise equal within a tolerance.
print(np.allclose([0.001, 0.5, 0.999], norm.cdf(vals)) )

# Generate random nums
r = norm.rvs(size=1000)

# Make histogram
ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
ax.set_xlim(x[0], x[-1])
ax.legend(loc='best', frameon=False)
plt.show()
fig.savefig("normaldist.png")
```

# Power Law Distribution #

```python
from scipy.stats import powerlaw
```

```python
# TODO: DISTRIBUTION FOR STARS LOC AND SCALE

fig, ax = plt.subplots(1,1)
a = 0.659
mean, var, skew, kurt = powerlaw.stats(a, moments='mvsk')

#  Display pdf
x = np.linspace(powerlaw.ppf(0.01, a), powerlaw.ppf(0.99, a), 100)
ax.plot(x, powerlaw.pdf(x,a), 'r-', lw=5, alpha=0.6, label='powerlaw pdf')

# Freeze distribution
rv = powerlaw(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

# Check accuracy of cdf and ppf
vals = powerlaw.ppf([0.001, 0.5, 0.999], a)
print(np.allclose(powerlaw.cdf(vals, a), [0.001, 0.5, 0.999]))

# Generate random numbers
r = rv.rvs(size=10000)

# Make histogram
ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
ax.set_xlim(x[0], x[-1])
ax.legend(loc='best')
plt.show()
fig.savefig("powerlaw.png")
```

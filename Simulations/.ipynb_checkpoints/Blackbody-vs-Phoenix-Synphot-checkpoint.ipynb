```python
from astropy.table import Table
from astropy.io import fits
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Blackbody imports 
from astropy.modeling.models import BlackBody
from astropy import units as u
from astropy import constants as const

from astropy.visualization import quantity_support
```

```python
dataBB = pd.read_csv('../CTIO/CTIO_ANDICAM.V_KPNO.txt', sep=' ', header=None, names=['Wavelength', 'Transmission'])

# USE WHOLE RANGE OF WAVELENGTHS

full_wav = np.arange(1000, 110000) * u.AA

# wavelength in Angstrom
# wavelengthsBB = dataBB['Wavelength'].to_numpy()*u.AA
wavelengthsBB = full_wav
# Probability of photon going through
transmission = dataBB['Transmission'].to_numpy()

pi = np.pi
# Flux for zero-point (Vega system)
F_m = (3640.33*u.Jy).cgs.value
# lambda ref in Angstroms
lambda_ref = 5443.85*u.AA
# R = (0.130 * u.Rsun).cgs.value
# Using radius of sun
R = (1*u.Rsun).cgs.value
d = (10*u.pc).cgs.value

# norm_integral = integrate.trapezoid(transmission/wavelengthsBB.cgs.value, x=wavelengthsBB.cgs.value)

T = 5772*u.K
bb = BlackBody(temperature=T)
B_nu = bb(wavelengthsBB).cgs.value
# B_v = integrate.trapezoid(B_nu.cgs.value*transmission/wavelengthsBB.cgs.value, x=wavelengthsBB.cgs.value)/norm_integral
B_nu_func = pi*B_nu
fig, ax = plt.subplots()
ax.plot(wavelengthsBB, B_nu_func)
ax.set_xlabel(r"$\lambda$" + "($\AA$)")
ax.set_ylabel("$\pi$" + " x " + r"$B_{\nu}$")
ax.set_xscale("log")
plt.show()
```

```python
# Opening fits file and reading data
hdul = fits.open("../Phoenix Model/phoenixm00_5700.fits")
hdul.info()
data = hdul[1].data

wavLen = data['WAVELENGTH']

# wavelength range from black body model
print(wavelengthsBB[0], wavelengthsBB[-1])

# Range of wavlengths used in black body model
# range_wavLen = (np.where(np.logical_and(wav_arr > 1000.0, wav_arr < 109999.0 ))[0])
# print(range_wavLen[0], range_wavLen[-1])


# f_lamda values for log(g) = 4.5
f_lambda = data['g45']

fig, ax = plt.subplots()

def f_lambda_func(wavelength_array):
    # Use 1e8 instead of 10**8 for non-integers
    return (f_lambda*1e8)*(wavelength_array*1e-8)**2/(const.c.cgs.value)
    
print(f_lambda_func(wavLen))

ax.plot(wavLen, f_lambda_func(wavLen))
ax.plot(wavelengthsBB, B_nu_func)

ax.set_xscale("log")
ax.set_xlabel("$\lambda$ ($\AA$)")
ax.set_ylabel(r"$F_{\nu}$")
# ax.set_yscale("log")
plt.show()
```
